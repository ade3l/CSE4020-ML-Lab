# K Means Clustering Algorithm

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = f'./driver_dataset.csv'
df = pd.read_csv(data,index_col="id")
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(df)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
df["cluster"] = kmeans.labels_
```

```python
sns.lmplot(data=df,x='mean_dist_day', y='mean_over_speed_perc', hue='cluster', palette='coolwarm', aspect=1, fit_reg=False)
```

```python
kmeans.score
```

```python
unique, counts = np.unique(kmeans.labels_, return_counts=True)
dict_data = dict(zip(unique, counts))
dict_data
```
